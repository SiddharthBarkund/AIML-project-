Linear datasets 

```python
from sklearn.datasets import load_iris
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

iris=load_iris()

x=iris.data
y=iris.target


scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)

sns.scatterplot(x=x_scaled[:,0],y=x_scaled[:,2],c=y)

dbscan = DBSCAN(
    eps=0.8,
    min_samples=5, 
    metric='euclidean',
    metric_params=None, 
    algorithm='auto',
    leaf_size=30,
    p=None,
    n_jobs=None
)

labels=dbscan.fit_predict(x_scaled)
```

Non- liner Data 

```python
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(
    n_samples=400, 
    shuffle=True, 
    noise=0.1,
    random_state=42
)
```

```python
x_scaled=scaler.fit_transform(x)

sns.scatterplot(
    x=x_scaled[:,0],
    y=x_scaled[:,1],
    hue=y
)
```

KMeans  ON NON-LINEAR DATASETS 

```python
from sklearn.cluster import KMeans
KMeans=KMeans(
    n_clusters=2,
    random_state=42

)
labels=KMeans.fit_predict(x_scaled)
```

```python
sns.scatterplot(
    x=x_scaled[:,0],
    y=x_scaled[:,1],
    c=labels
)
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(
    eps=0.5,
    min_samples=5
)

labels = dbscan.fit_predict(x_scaled)
```

```python
sns.scatterplot(
    x=x_scaled[:,0],
    y=x_scaled[:,1],
    c=labels
)
```
